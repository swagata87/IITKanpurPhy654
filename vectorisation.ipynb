```python
import time
import numpy
import array
```

Example 1

Screenshot from 2024-08-21 13-19-46.png

```python
# 8 bytes int
a = array.array('q')
for i in range(100000):
    a.append(i);

b = array.array('q')
for i in range(100000, 200000):
    b.append(i)
```

```python
len(a), len(b)
```

```python
# Method 1: without vectorization

tic = time.process_time() # start the timer in stopwatch
dot = 0.0;

for i in range(len(a)):  # using a for loop
      dot += a[i] * b[i]

toc = time.process_time() # end the timer

print("dot_product = ", dot);
print("Computation time = " , 1000*(toc - tic ) , " ms")

# run multiple times, and you will see different  values (but roughly in same order of magnitude)
```

```python
# Method 2: with vectorization (no for loop)

new_tic = time.process_time()
new_dot_product = numpy.dot(a, b)
new_toc = time.process_time()

print("\nnew_dot_product = " , new_dot_product)
print("Computation time = " , 1000*(new_toc - new_tic ) , " ms\n\n")
```

Example 2

Screenshot from 2024-08-21 13-42-34.png

```python
a = array.array('i')
for i in range(200):
    a.append(i);

b = array.array('i')
for i in range(200, 400):
    b.append(i)
```

```python
len(a), len(b)
```

```python
tic = time.process_time()
outer_product = numpy.zeros((200, 200))

for i in range(len(a)):
   for j in range(len(b)):
      outer_product[i][j]= a[i]*b[j]

toc = time.process_time()

print("outer_product = " , outer_product)
print("Computation time = " , 1000*(toc - tic ) , " ms")
```

```python
new_tic = time.process_time()
new_outer_product = numpy.outer(a, b)
new_toc = time.process_time()

print("outer_product = " , new_outer_product)
print("\nComputation time = " , 1000*(new_toc - new_tic ) , " ms\n\n")
```
